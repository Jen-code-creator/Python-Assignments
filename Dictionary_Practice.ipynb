```python
# Single Key is always used
student = {"name":"krish","age":32,"name":24}
print(student)
```

```python
## Accessing Dictionary Elements
student = {"name":"krish", "age":32,"grade":'A'}
print(student)
```

```python
## Accessing Dictionary elements
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
## Modifying Dictionary Elements
## Dictionary are Mutable, so you can add, update or delete elements
print(student)
```

```python
## update value for the key
student["age"]=33   
print(student)
## Added a new key and value 
student["address"]="India"
print(student)
```

```python
## Delete the key and value pairs
del student['grade']
print(student)
```

```python
print(student)
```

```python
## Dictionary Methods
## get all the keys
keys = student.keys()
print(keys)
## get all the values
values = student.values()
print(values)

## get all key value pairs
items = student.items()
print(items)
```

```python
## Shallow Copy
student_copy = student
print(student_copy)
print(student)
```

```python
student["name"]="Krish1"
print(student_copy)
print(student)
```

```python
## Shallow Copy
student_copy1=student.copy()
print(student_copy1)
print(student)
```

```python
student["name"]="Krish3"
print(student_copy1)
print(student)
```

```python
## Iterating over Dictionaries
## You can use use loops to iterate over dictionaries, keys, values or items

## Iterating over Keys
for keys in student.keys():
    print(keys)
```

```python
## Iterating over the values
for values in student.values():
    print(values)
```

```python
## Iterate over key value pairs
for items in student.items():
    print(items)

for values,keys in student.items():
    print(f"{Keys}:{Values}")
```

```python
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries
students = {
    "student1":{"name":"krish","age":32},
    "student2":{"name":"peter","age":"23"}
}
print(students)
```

```python
## Access Nested Dictionaries elements
print(students["student1"]["name"])
print(students["student2"]["age"])
```

```python
## Iterating over Nested Dictionaries
for student_id, student_info in student.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Iterating over Nested Dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension
squares={x:x**2 for x in range (5)}
print(squares)
```

```python
## Conditional Dictionary Comprehension
even={x:x**2 for x in range (10) if x%2==0}
print(even)
```

```python
## Practical Examples
## Use a Dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
